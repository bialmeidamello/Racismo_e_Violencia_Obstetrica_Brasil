##Racismo e Violência obstétrica no Brasil
“A violência obstétrica tem implicações sobre a morbimortalidade materna. Dentre os motivos estão: (a) o manejo agressivo do parto vaginal; (b) o constrangimento ou coerção à cesárea; (c) a negligência em atender mulheres que expressam seu sofrimento ou que pedem ajuda de modo insistente; (d) a hostilidade contra maneiras de assistência que fogem do modelo hegemônico – o uso de doulas, por exemplo; (e) a hostilidade, negligência e retardo em dar atendimento à mulheres que sofreram aborto por acreditar que esse foi provocado; (f) o impedimento à presença de um acompanhante”. (Pág. 8 - MASSON, Marcelo M. - 2017)

## <font color='pink'> **Tratando os dados** </font>
Dados coletados pela PNS (Pesquisa Nacional de Saúde) nos anos de 2013 e 2019


1) Tratando dados relacionados a porcentagem de mulheres que tiveram o último parto realizado no estabelecimento de saúde indicado durante o pré-natal


```python
parto_hosp_indicado = pd.read_csv('/content/realizaram_parto_hospital_indicado.csv', sep = ';', usecols=['Ano','Nome','Valor'])

parto_hosp_indicado.rename(columns={'Nome': 'Raça/Etnia', 'Valor': 'Porcentagem'}, inplace=True)

parto_hosp_indicado['Porcentagem'] = parto_hosp_indicado['Porcentagem'].astype(str) + '%'

parto_hosp_indicado
```

```python
parto_hosp_indicado.to_csv('parto_hosp_indicado_final.csv', index = False)
parto_hosp_indicado_final = pd.read_csv('/content/parto_hosp_indicado_final.csv')
parto_hosp_indicado_final
```
